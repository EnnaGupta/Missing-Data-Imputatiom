```python
data = pd.read_csv(r"/content/GAN_original.csv")
```

```python
dt.loc[dt['Crop'].isna()]
dt_new= dt
```

```python
from numpy.random import randint
values = randint(0, 30000, 50)
values = [*set(values)]
values
```

```python
dt_new.loc[values,'YEILD/AREA']='NaN'
```

```python
dt_new.loc[dt_new['YEILD/AREA'] == 'NaN']
```

```python
from sklearn.impute import KNNImputer
k =2   # Number of neighbors to consider
imputer = KNNImputer(n_neighbors=k)

# Perform KNN imputation on the dataset
imputed_data = imputer.fit_transform(dt_new)

# Convert the imputed data back to a DataFrame
imputed_data = pd.DataFrame(imputed_data, columns=dt_new.columns)

# Print the imputed dataset
imputed_data
```
