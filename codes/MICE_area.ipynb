MICE


```python
#creating dataset for MICE
MICEdf = msgdata.drop(data.index[0:8258])
MICEdf =MICEdf[['YEILD/AREA','Area','Season','Crop']]
```

```python
MICEdf[['YEILD/AREA','Area','Season','Crop']] = MICEdf[['YEILD/AREA','Area','Season','Crop']].astype('category')
```

```python
MICEdf =MICEdf.loc[MICEdf['Crop'] == 'Paddy']
```

```python
#missing dataset
MICEdf.loc[MICEdf['YEILD/AREA'].isna()]
```

```python
import miceforest as mf
```

```python
# Create kernel.
kds = mf.ImputationKernel( MICEdf, save_all_iterations=True,  random_state=100)

# Run the MICE algorithm for 2 iterations
kds.mice(2)

# Return the completed dataset.
df_imputed = kds.complete_data()
```

```python
MICEdf.head()
```

```python
df_imputed.head()
```

```python
#increasing iteration
kds.mice(iterations=5, n_estimators=50)
```

```python
df_imputed2 = kds.complete_data()
df_imputed2.loc[df_imputed2['YEILD/AREA'].isna()]
```

```python
df.loc[df['YEILD/AREA'].isna()]
```

```python
extracted_col = df["YEILD/AREA"]
extracted_col
```

```python
df_imputed2.insert(1, "org_Y/A", extracted_col)
display(df_imputed2)
```

```python
Array1 =df_imputed2["YEILD/AREA"].to_numpy()
Array2 = df_imputed2["org_Y/A"].to_numpy()
```

```python
from sklearn import metrics
meanAbErr = metrics.mean_absolute_error(Array1, Array2)
meanSqErr = metrics.mean_squared_error(Array1, Array2)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(Array1, Array2))

#print('R squared: {:.2f}'.format(mlr.score(x,y)*100))
print('Mean Absolute Error:', meanAbErr)
print('Mean Square Error:', meanSqErr)
print('Root Mean Square Error:', rootMeanSqErr)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
error = mean_absolute_percentage_error(Array1, Array2)
print(' %Error:', error)
```
