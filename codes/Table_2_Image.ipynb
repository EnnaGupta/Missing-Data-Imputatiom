```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#data loading of year 2013-14 and for crop paddy without 0 yield/Area value
data = pd.read_csv(r"/content/tab2img.csv",low_memory=False)
df = data
```

```python
#deleting first column
df.drop(data.columns[0], axis=1, inplace=True)
df.head()
```

```python
#missing values =0
df=df.fillna(0)
df
```

```python
from PIL import Image
import numpy as np
```

```python
#1
# Convert data to a NumPy array
data_array = df.values

# Normalize the data to [0, 1] range
normalized_data = (data_array - np.min(data_array)) / (np.max(data_array) - np.min(data_array))

# Reshape the data to match the desired image dimensions
image_width = 77 # Adjust this value based on the desired image width
image_height = 91  # Adjust this value based on the desired image height
reshaped_data = normalized_data.reshape(-1, image_width, image_height)

# Create an empty image to store the tabular data
data = np.zeros((image_width, image_height))
#data = np.zeros((image_width, image_height, 3), dtype=np.uint8)
#assigning array
for i in range(image_width):
    for j in range(image_height):
           data[i, j] = reshaped_data[0,i,j]



#data[0:256, 0:256] = [255, 0, 0] # red patch in upper left
img = Image.fromarray(data, 'RGB')
img.save('my.png')
img.show()
#bt with nan values also zero & can we convert normalised missing values to real
```
