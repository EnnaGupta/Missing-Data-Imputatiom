```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
imputer = IterativeImputer(random_state=100, max_iter=10)
```

```python
import pandas as pd
orgdf = pd.read_csv(r"/content/org_data.csv",low_memory=False)
orgdf.head()
```

```python
orgdata = orgdf.loc[:, ['YEILD/AREA','Lattitude..167','longitude','Season','Crop']]
```

```python
misdf = pd.read_csv(r"/content/missing_data.csv",low_memory=False)
misdf.head()
```

```python
misdf =misdf[['YEILD/AREA','Lattitude..167','longitude','Season','Crop']]
```

```python
new_df = misdf.copy()
for col in misdf.select_dtypes(include='object').columns:
    new_df = pd.get_dummies(new_df, columns=[col])
misdf = new_df
misdf.head()
```

```python
misdf_train = misdf.loc[:, ['YEILD/AREA','Lattitude..167','longitude','Season','Crop_Paddy']]
misdf_train.loc[misdf_train['YEILD/AREA'].isna()]
```

```python
import miceforest as mf
```

```python
# Create kernel.
kds = mf.ImputationKernel(
  misdf_train,
  save_all_iterations=True,
  random_state=100
)

# Run the MICE algorithm for 2 iterations
kds.mice(2)

# Return the completed dataset.
df_imputed = kds.complete_data()
```

```python
kds.mice(iterations=5, n_estimators=50)
```

```python
df_imputed2 = kds.complete_data()
df_imputed2.head()
```

```python
extracted_col = orgdf["YEILD/AREA"]
display(extracted_col)
```

```python
df_imputed2.insert(1, "imp_Y/A", extracted_col)
display(df_imputed2)
```

```python
Array1 =df_imputed2["YEILD/AREA"].to_numpy()
Array2 = df_imputed2["imp_Y/A"].to_numpy()
```

```python
import numpy as np
from sklearn import metrics
meanAbErr = metrics.mean_absolute_error(Array1, Array2)
meanSqErr = metrics.mean_squared_error(Array1, Array2)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(Array1, Array2))
#print('R squared: {:.2f}'.format(mlr.score(x,y)*100))
print('Mean Absolute Error:', meanAbErr)
print('Mean Square Error:', meanSqErr)
print('Root Mean Square Error:', rootMeanSqErr)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
error = mean_absolute_percentage_error(Array1, Array2)
print(' %Error:', error)
```
